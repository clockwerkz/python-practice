```python
#This works to find unique values of a column but is messy
list(set(df['region'].values.tolist()))
```

```python
#cleaner way to get unique values:
df['region'].unique()
```

```python
graph_df = pd.DataFrame()
for region in df['region'].unique()[:16]:
    print(region)
    region_df = df.copy()[df["region"]=='Albany']
    region_df.set_index('Date', inplace=True)
    region_df.sort_index(inplace=True)
    region_df[f'{region}price25ma'] = region_df['AveragePrice'].rolling(25).mean()
    
    if graph_df.empty:
        graph_df = region_df[[f'{region}price25ma']]
        
    else:
        graph_df = graph_df.join(region_df[f'{region}price25ma'])
```

```python
import pandas as pd

df = pd.read_csv('data/avocado.csv')
df = df.copy()[df['type'] == 'organic']
df['Date'] = pd.to_datetime(df['Date'])

df.sort_values(by="Date", ascending=True, inplace=True)
```

```python
graph_df = pd.DataFrame()
for region in df['region'].unique()[:16]:
    print(region)
    region_df = df.copy()[df["region"]=='Albany']
    region_df.set_index('Date', inplace=True)
    region_df.sort_index(inplace=True)
    region_df[f'{region}price25ma'] = region_df['AveragePrice'].rolling(25).mean()
    
    if graph_df.empty:
        graph_df = region_df[[f'{region}price25ma']]
        
    else:
        graph_df = graph_df.join(region_df[f'{region}price25ma'])
        
graph_df.tail()
```

```python
graph_df.dropna().plot(figsize=(8,5))
```
